тижневу підписку. Підписка коштує 9.99 за тиждень. Юзер може відписатись у<br>

Ваше завдання - розрахувати LTV за допомогою Python. Будемо оцінювати<br>
ваше загальне розуміння завдання та результат. У файлі data_analytics.csv ви<br>
знайдете фрагмент бази даних, що вам потрібна, де:<br>

Event Date – дата коли відбулась подія (тріал чи продовження підписки).<br>

Subscription Offer Type – тип підписки (якщо вказано Free Trial - значить це<br>
початок тріалу, якщо поле пусте - це підписка).<br>

LTV – lifetime value. У цьому випадку рахується через конверсії з тріалу в<br>
оплату і потім в наступні оплати. <br>

LTV = ([1]+[2]+[3]+[4]+[5]), де [1] = (dev proceeds * Conversion to trial), а [2] = [1] * 1st
purchase to 2nd, [3] = [2] * 2nd to 3rd і тд.<br>

Важливо! Apple Store забирає собі 30% від вартості підписки. LTV в цьому<br>
випадку ми будемо рахувати на 6 тижнів(1 тиждень тріалу та 5 підписки).<br>

LTV = [1] + [2] + [3] + [4] + [5]  
[1] = (dev_proceeds * conversion_to_trial)  
[2] = [1]* 1st_to_2nd  
[3] = [2]* 2nd_to_3d  
[4] = [3]* 3d_to_4th  
[5] = [4]* 4th_to_5th

### Згрупуємо по 'user_id' та застосуємо rank до 'purchase_date', з метою виділити порядковий номер кожного івенту для юзера

```python
df['rank_purch'] = df.groupby('Subscriber ID')['Event Date'].rank(ascending = True, method = 'first')
```

```python
df
```

```python
users = df.groupby('rank_purch')['Subscriber ID'].count().reset_index()
```

```python
users_copy = users.copy()
```

```python
users_copy
```

###  Посчитаем конверсию 

```python
users_copy['rank_purch'] = users_copy['rank_purch'] - 1
```

```python
users = users.merge(users_copy, on = 'rank_purch')
```

```python
users
```

```python
users['conv']= users['Subscriber ID_y']/users['Subscriber ID_x']
```

```python
users = users.drop(columns=['Subscriber ID_x', 'Subscriber ID_y'])
```

```python
users_copy = users_copy.merge(users, on = 'rank_purch' , how = 'left')
```

```python
users_copy
```

### Посчитаем LTV

```python
dev_proceeds = 1019 * 9.99
first = float(dev_proceeds * users_copy[users_copy['rank_purch']==1]['conv'])
second  = float(first * users_copy[users_copy['rank_purch']==2]['conv'])
third = float(second * users_copy[users_copy['rank_purch']==3]['conv'])
fourth = float(third * users_copy[users_copy['rank_purch']==4]['conv'])   
fifth = float(fourth * users_copy[users_copy['rank_purch']==5]['conv']) 
LTV = first + second  + third + fourth + fifth 
```
